```python
from pyspark.sql.functions import col, when

greater_accra_aliases = [
    "Greater Accra",
    "Accra",
    "Accra North",
    "Accra East",
    "Ga East Municipality",
    "Ga East Municipality, Greater Accra",
    "Shai Osudoku District, Greater Accra",
    "Ledzokuku-krowor",
    "East Legon",
    "Tema West Municipal"
]

df = spark.table("med_atlas_ai.default.facility_records")

df_cleaned = df.withColumn(
    "state",
    when(col("state").isin(greater_accra_aliases), "Greater Accra").otherwise(col("state"))
)

# Write the cleaned data back to the table
df_cleaned.write.mode("overwrite").saveAsTable("med_atlas_ai.default.facility_records")

print("Table updated successfully!")
```

```python
from pyspark.sql.functions import col, when

# Define aliases for each region
ashanti_aliases = [
    "Ashanti",
    "Asokwa-kumasi",
    "Ejisu Municipal",
    "Ahafo Ano South-east"
]

western_aliases = [
    "Western",
    "Takoradi"
]

western_north_aliases = [
    "Western North"
]

central_aliases = [
    "Central",
    "Central Ghana",
    "Keea"
]

eastern_aliases = [
    "Eastern"
]

volta_aliases = [
    "Volta",
    "Central Tongu District"
]

oti_aliases = [
    "Oti"
]

northern_aliases = [
    "Northern"
]

north_east_aliases = [
    "North East"
]

savannah_aliases = [
    "Savannah"
]

upper_east_aliases = [
    "Upper East"
]

upper_west_aliases = [
    "Upper West",
    "Sissala West District"
]

bono_aliases = [
    "Bono",
    "Dormaa East"
]

bono_east_aliases = [
    "Bono East",
    "Techiman Municipal"
]

ahafo_aliases = [
    "Ahafo",
    "Asutifi South"
]

df = spark.table("med_atlas_ai.default.facility_records")

df_cleaned = df.withColumn(
    "state",
    when(col("state").isin(ashanti_aliases), "Ashanti")
    .when(col("state").isin(western_aliases), "Western")
    .when(col("state").isin(western_north_aliases), "Western North")
    .when(col("state").isin(central_aliases), "Central")
    .when(col("state").isin(eastern_aliases), "Eastern")
    .when(col("state").isin(volta_aliases), "Volta")
    .when(col("state").isin(oti_aliases), "Oti")
    .when(col("state").isin(northern_aliases), "Northern")
    .when(col("state").isin(north_east_aliases), "North East")
    .when(col("state").isin(savannah_aliases), "Savannah")
    .when(col("state").isin(upper_east_aliases), "Upper East")
    .when(col("state").isin(upper_west_aliases), "Upper West")
    .when(col("state").isin(bono_aliases), "Bono")
    .when(col("state").isin(bono_east_aliases), "Bono East")
    .when(col("state").isin(ahafo_aliases), "Ahafo")
    .otherwise(col("state"))
)

# Write the cleaned data back to the table
df_cleaned.write.mode("overwrite").saveAsTable("med_atlas_ai.default.facility_records")

print("Table updated successfully!")
print(f"Standardized all 16 regions of Ghana")
```
